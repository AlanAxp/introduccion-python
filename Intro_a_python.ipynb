_**Funciones y lambdas**_

```python
# Las funciones se definen escribiendo el nombre de la función y posteriormente los argumentos

def nombre_funcion(argumento):
    
    # facer algo dentro de la función
    
    # despues sigue el return que puede regresar lo que quremos que regrese al llamar la func. 
    # En este caso solo queremos regresar el argumento... Que funcion tan aburrida...

    return argumento


# Ejemplo praáctico de función
    
def Info_alumno(califs):
    return sum(califs)/len(califs)

# sum(algo_iterable) suma los elementos del arreglo
# len(algo_iterable) nos da la longitud del arreglo, su "cardinalidad".
 
# Esta pequeña función lo que hace es calcular el promedio de las califaciones de algún 
# alumno. 

# Para usar la función que acabamos de crear

lista_calif = [ 10, 10, 4, 6, 7, 8, 10, 5 ]
print("La calificación del alumno es de", Info_alumno(lista_calif))

# Al momento de ingresar los argumentos de la función no es necesario que tenga el mismo 
# nombre que con la que se definió el argumento en la función.
```

### Calcular PI con el método de Montecarlo

```python
#Aproximaremos el valor de pi con el metodo de MonteCarlo
def pi_mc_2d(shoots):
    
    #Fijemos un valor inicial para la suma de la cantidad de veces que hemos caido dentro del circulo
    inicio = 0.0
    
    #Queremos que se repita tantas veces como el parámetro shoots que pasamos
    for i in range(shoots):
        
        # Creamos un punto x, y aleatorio en el intervalo 0,1
        x = rand()
        y = rand()
        
        # Para considerar que un punto esta en el circulo de radio 1 tiene que cumplir...
        if x*x+y*y <= 1 :
            
            # Esto incrementa en uno si le hemos atinado al interior del circulo
            inicio = inicio + 1
    
    # Queremos el ratio de puntos dentro 
    return (inicio / shoots)*4

def lista_pi(A,N): # A es la long de lista que quermos, N = puntos arrojados al cuadradp
    pis = []
    for i in range(A):
        
        #Guardamos la repeticion del experimento para poder hacer estadistica
        pis.append(pi_mc_2d(N))
    return pis

def prom_var_err(DAT):
    prom = 0.0
    cuadrados = 0.0
    N = len(DAT)
    for i in range(N):
        prom = prom + DAT[i]
        cuadrados = cuadrados + DAT[i]*DAT[i]    
    xbar = prom/N #    Valor del promedio
    var = (cuadrados - N*xbar*xbar)/(N-1)  #Aquí calculamos la varianza
    Er = np.sqrt(var/N)
    return xbar,var,Er
```

```python
#Ejecutando nuestro programa...


from time import time #importamos la función time para capturar tiempos

tiempo_inicial = time()  #Para calcular el tiempo de computo

Pis = lista_pi(1000,40000); #bajar parametros para que su compu no tarde mucho
result = prom_var_err(Pis)

tiempo_final = time()
tiempo_ejecucion = tiempo_final - tiempo_inicial

print("El valor aprox de π es", result[0],"con un error estandar de", result[2])
print( 'Con un tiempo de ejecucion de',tiempo_ejecucion, "segundos") #En segundos
```

## Programación orientada a objetos con Python

```python
class pelotita: 
    
    # Iniciamos el constructor con los valores iniciales
    def __init__(self, y0, Vy0, m0):
        self.y = y0
        self.Vy = Vy0 
        self.m = m0
        
        self.g = 9.8
        self.Fy = 0.0
        
    # Creamos el método para calcular el peso
    def calcularFuerza(self):
        self.Fy = - self.m * self.g
        
    # Creamos el método que describe el movimiento
    def muevete(self, dt):
        self.y += self.Vy * dt
        self.Vy += dt * self.Fy / self.m
```

```python
# Instanciar el objeto, con los parametros definidos en el constructor
bolita = pelotita(0.0, 3.0, 5)

# Llamamos al médoto para calcular el peso
bolita.calcularFuerza()

# Creamos una iteracion para un intervalo de tiempo 
t = 0.0
Delta_t = 0.01
T = []
Y = []

while t < 1.0:
    bolita.muevete(Delta_t)
    t += Delta_t
    T.append(t)
    Y.append(bolita.y)
```

```python
import matplotlib.pyplot as plt

plt.plot(T, Y)
plt.ylabel("posicion en y")
plt.xlabel("tiempo transcurrido")
```
